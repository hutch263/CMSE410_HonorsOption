```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataframe = pd.read_csv('Noncommunicable_diseases.csv')
percent_from_NCDs = my_dataframe[my_dataframe["Indicator Name"] == 'Percentage of total deaths due to NCDs']

import seaborn as sns
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=percent_from_NCDs, kind="bar",
    x="Region", y="Numeric", hue="Gender",
     palette="dark", alpha=.6, height=6, aspect = 2.7
).set(title='Percent of Deaths Due to NCDs by Global Region and Gender')
plt.yticks()
g.despine(left=True)
g.set_axis_labels("Region", 'Percent of Deaths Due to NCDs')
g.legend.set_title("Gender")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataframe = pd.read_csv('Noncommunicable_diseases.csv')
percent_from_NCDs = my_dataframe[my_dataframe["Indicator Name"] == 'Percentage of total deaths due to NCDs']

import seaborn as sns
sns.set_theme(style="whitegrid")

sns.set(font_scale=1.25)
g = sns.catplot(
    data=percent_from_NCDs, kind="bar",
    x="Region", y="Numeric", hue="Gender",
     palette="dark", alpha=.6, height=6, aspect = 2.7
).set(title='Percent of Deaths Due to NCDs by Global Region and Gender')

plt.suptitle("Error bars represent a 95% confidence interval", x=0.49, y=0, ha='center', fontsize='medium')
plt.yticks()
g.despine(left=True)
g.set_axis_labels("Region", 'Percent of Deaths Due to NCDs')
g.legend.set_title("Gender")

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.suptitle.html#matplotlib.pyplot.suptitle
#  https://stackoverflow.com/questions/25328003/how-can-i-change-the-font-size-using-seaborn-facetgrid 
```

```python
df= pd.read_csv('HospInfo.csv')
```

```python
df = df[df['Hospital overall rating'] != 'Not Available']
df.head()
```

```python
df.loc[df['Hospital Ownership'] == 'Government - Hospital District or Authority', 'Hospital Ownership'] = 'Government HD/A'
df['Hospital overall rating'] = df['Hospital overall rating'].astype(float)
# Chat GPT 3.5: how to change one of the category names
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Hospital Ownership', y='Hospital overall rating', data=df)
plt.title('Ratings of Hospitals with Different Ownerships')
plt.xlabel('Hospital Ownership')
plt.ylabel('Overall Hospital Rating')

plt.xticks(rotation=45, ha='right')

plt.show()


#ChatGPT 3.5: how to make plt boxplots from categories in column and values in another
```

```python
# 202 students may not choose choropleth maps yet but I wanted to do a better practice for me

import plotly.express as px


avg_values = df.groupby('State')['Hospital overall rating'].mean().reset_index()


fig = px.choropleth(df, locations='State', 
                    locationmode='USA-states',
                    color='Hospital overall rating',
                    color_continuous_scale="Viridis",
                    range_color=(0, 5),
                    scope="usa",
                    labels={'Hospital overall rating':'Average Overall Hospital Rating'}
                          )

fig.update_layout(
    coloraxis_colorbar=dict(
        title='',
        tickfont=dict(size=14, color='black')))

fig.update_layout(
    title_text='<b>Average Hospital Ratings for Each State</b>',
    title_font=dict(size=25,color='black'),
    title_x=0.5,  # Center the title horizontally
    title_y=0.92) # Adjust the vertical position of the title

fig.update_layout(
    width=800,
    height=500,  
)


fig.show()
```
